Mary, Agatha and Noel work with Zenith Bank Nigeria. Mary works in the Retail Banking Division, Agatha works in the Global Banking Division and Noel works in the Commercial Banking Division. The three divisions have some unique services and some mutual services as indicated:

#### Retail Banking:
<ul><li>Lines of credit</li><li>
Investment management and accounts</li><li>
Insurance</li><li>
Retirement and education accounts</li><li>
    loans and mortgagges</li><li>
    Checking and saving</li></ul>

#### Global Banking:
<ul><li>Multi-currency management services and products</li><li>
Foreign currency accounts</li><li>
Foreign currency credit cards</li><li>
Transborder advisory services</li><li>
Liquidity management</li></ul>

#### Commercial Banking:
<ul><li>Lines of credit</li><li>
Investment management and accounts</li><li>
Insurance</li><li>
Advisory services</li></ul>

With your knowledge in OOP develop a python program that will take as input an employee name and division, and then displays the service rendered in the division. The program should highlght key concepts of OOP; class objects inheritance and polymorphism.

#### Hints:
<ul><li>Create parent class <b>zenith()</b> with two methods <b>unique_services()</b> and <b>mutual_services()</b></li><li>
    The different divisions can be subclasses of the parent class, inheriting the parent methods.</li><li>
    Ploymorphism can be used to overide exclusive services.</li><ul>

```python
class Zenith:
    def __init__(self,name,division):
        self.name = name
        self.division = division
    def unique_services(self):
        pass
    def mutual_services(self):
        pass
    
class Retail(Zenith):
    def __init__(self,name,division):
        Zenith.__init__(self,name,division)
        
    #A tuple of the services offered by retail banking
    UServices = ["Retirement and education accounts","Loans and Mortgages","Checking and saving"]
    MServices = ["Lines of credit","Investment management and accounts","Insurance"]
    
    def unique_services(self):
        print("The unique services of Retail Banking are:",self.UServices)
    def mutual_services(self):
        print("The mutual services that Retail Banking has with the other divisions are:",self.MServices)
        
class Global(Zenith):
    def __init__(self,name,division):
        Zenith.__init__(self,name,division)
        
    #A tuple of the services offered by global banking
    UServices = ["Multi-currency management services and products", "Foreign currency accounts","Foreign currency credit cards","Transborder advisory services","Liquidity management"]
    MServices = []
    
    def unique_services(self):
        print("The unique services of Global Banking are:",self.UServices)
    def mutual_services(self):
        print("Global Banking has no mutual services with the other divisions")
class Commercial(Zenith):
    def __init__(self,name,division):
        Zenith.__init__(self,name,division)
        
    #A tuple of the services offered by commercial banking
    UServices = ["Advisory services"]
    MServices = ["Lines of credit","Investment management and accounts","Insurance"]
    
    def unique_services(self):
        print("The unique services of Retail Banking are:",self.UServices)
    def mutual_services(self):
        print("The mutual services that Retail Banking has with the other divisions are:",self.MServices)

name = str(input("What is your name?"))
division = str(input("What is your division?"))

if division == "Retail":
    user = Retail(name,division)
    user.unique_services()
    user.mutual_services()
elif division == "Global":
    user = Global(name,division)
    user.unique_services()
    user.mutual_services()
elif division == "Commercial":
    user = Commercial(name,division)
    user.unique_services()
    user.mutual_services()
```
