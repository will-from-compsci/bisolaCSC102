## Class Project I

Develop a python OOP program that creates an abstract base class called coup_de_ecriva.  The base class will have one abstract method called <b>Fan_Page</b> and four subclassses namely; <b>FC_Cirok, Madiba_FC, Blue_Jay_FC and TSG_Walker</b>. The program will receive as input the name of the club the user supports and instantiate an object that will invoke the <b>Fan_Page</b> method in the subclass that prints Welcome to <b>"club name"</b>.

<p><b>Hint:</b></p>
The subclasses will use <b>Single Inheritance</b> to inherit the abstract base class.
 

```python
from abc import ABC, abstractmethod

class coup_de_ecriva(ABC):
    @abstractmethod
    def Fan_Page(self):
        pass

class FC_Cirok(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Cirok")

class Madiba_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Madiba")

class Blue_Jay_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Blue Jay")

class TSG_Walker(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to TSG")

club = input("What name of your  club?")
if club == "Cirok":
    obj = FC_Cirok()
    obj.Fan_Page()
elif club == "Madiba":
    obj = Madiba_FC()
    obj.Fan_Page()
elif club == "Blue Jay":
    obj = Blue_Jay_FC()
    obj.Fan_Page()
elif club == "TSG":
    obj = TSG_Walker()
    obj.Fan_Page()
else:
    print("Please choose a valid club name")
```

<p><b>Hints:</b></p>
    <ul><li>The abstract base class is called <b>External_Vendors()</b>.</li><li>
        The abstract method is called <b>menu()</b>.</li><li>
The subclasses (the different vendors) will inherit the abstract base class.</li><li>
        Each subclass will have a normal method called <b>menu()</b>.</li></ul>
    
       

```python
from abc import ABC, abstractmethod
import pandas as pd

cooperative_menu = {'Main Meal':['Jollof Rice and Stew','White Rice and Stew','Fried Rice','Salad','Plantain'],
                    'Price(N)': [200,200,200,100,100]}

faith_menu = {'Main Meal':['Fried Rice','White Rice and Stew','Jollof Rice','Beans','Chicken'],
                    'Price(N)': [400,400,400,200,1000]}

student_centre_menu = {'Main Meal':['Chicken Fried Rice','Pomo Sauce','Spaghetti Jollof','Amala/Ewedu','Semo with Eforiro Soup'],
                    'Price(N)': [800,300,500,500,500]}

class External_Vendors(ABC):
    @abstractmethod
    def menu(self):
        pass

class Cooperative(External_Vendors):
    def menu(self):
        df = pd.DataFrame(cooperative_menu)
        return df

class Faith(External_Vendors):
    def menu(self):
        df = pd.DataFrame(faith_menu)
        return df

class Student_Centre(External_Vendors):
    def menu(self):
        df = pd.DataFrame(student_centre_menu)
        return df

vendor = input("What name of your desired vendor?")
if vendor == "Cooperative":
    obj = Cooperative()
    obj.menu()
elif vendor == "Faith":
    obj = Faith()
    obj.menu()
elif vendor == "Student Centre":
    obj = Student_Centre()
    obj.menu()
else:
    print("Please choose a valid name")
```
